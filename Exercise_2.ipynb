```python
### Programming task 1- Laplace Edge detector

import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

path= os.getcwd()
Img_path = os.path.join(path, 'images/img2.png')

img = cv2.imread(Img_path)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_smooth = cv2.GaussianBlur(img_gray, (3,3), 0)

kernel_size = 3
ddepth = cv2.CV_16S
laplacian = cv2.Laplacian(img_smooth, ddepth, ksize=kernel_size)
laplacian_abs = cv2.convertScaleAbs(laplacian)

plt.subplot(121)
plt.imshow(img);
plt.axis("off")
plt.title('Original image')

plt.subplot(122)
plt.imshow(laplacian_abs,cmap = 'gray');
plt.axis("off")
plt.title('Laplacian edge');

#Save image (Uncomment the line to save)
#plt.savefig("./Results/Result2.png",dpi=300,bbox_inches='tight')
```

```python
### Programming task 2- SIFT feature detection

import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

path= os.getcwd()
Img1_path = os.path.join(path, 'images/img1.png')
Img2_path = os.path.join(path, 'images/img2.png')

#Canny edge Detection
img = cv2.imread(Img1_path)
img1 = cv2.imread(Img2_path)

img_copy = img.copy()
img_copy_gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)

img1_copy = img1.copy()
img1_copy_gray = cv2.cvtColor(img1_copy, cv2.COLOR_BGR2GRAY)


sift=cv2.SIFT_create()

kp1, des1 = sift.detectAndCompute(img_copy_gray, None)
kp2, des2 = sift.detectAndCompute(img1_copy_gray, None)

img_kp1=cv2.drawKeypoints(img_copy_gray,kp1,None,[255,0,0],flags=None)

img_kp2=cv2.drawKeypoints(img1_copy_gray,kp2,None,[255,0,0],flags=None)

bf = cv2.BFMatcher()
matches=bf.match(des1,des2)

matches = sorted(matches, key = lambda x:x.distance)

result=cv2.drawMatches(img_copy_gray,kp1,img1_copy_gray,kp2,matches[:min(200,len(matches))],None,[0,255,0],flags=2)

plt.subplot(221)
plt.imshow(img_kp1)
plt.axis("off")
plt.tight_layout(pad=1.0)
plt.title('Image1')

plt.subplot(222)
plt.imshow(img_kp2)
plt.axis("off")
plt.tight_layout(pad=1.0)
plt.title('Image2')

plt.subplot(212)
plt.imshow(result)
plt.axis("off")
plt.tight_layout(pad=2.0)
plt.title('Feature Matching')

#Save image (Uncomment the line to save)
#plt.savefig("./Results/Result4.png",dpi=300,bbox_inches='tight')
```

```python
##############
# Problem 2
##############

### Programming task 1- Step-by-step implementation of Canny edge detector

import cv2
import os
import numpy as np
from matplotlib import pyplot as plt

path= os.getcwd()
Img_path = os.path.join(path, 'images/image.png')

img = cv2.imread(Img_path)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def gaussian_kernel(size,std_dev):
    size=int(size)
    x,y=np.mgrid[-size:size+1,-size:size+1]
    normal = 1 / (2.0 * np.pi * std_dev**2)
    g_kernel = np.exp(-((x**2+y**2)/(2*std_dev**2)))*normal
    return g_kernel

def sobel_filter(img):
    Sx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]],np.float32)
    Sy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]],np.float32)
     
    sobel_x = cv2.filter2D(img, -1, Sx)
    sobel_y = cv2.filter2D(img, -1, Sy)
    
    magnitude = np.hypot(sobel_x,sobel_y)
    magnitude = magnitude*255/magnitude.max()
    magnitude = magnitude.astype('uint8')
    angle = np.arctan2(sobel_y,sobel_x)
    
    return magnitude,angle

def non_max_supression(img,ang):
    w,h = img.shape
    blank_img = np.zeros((w,h), dtype=np.uint8)
    ang = ang*180./np.pi
    ang[ang < 0] += 180
    
    for i in range(1,w-1):
        for j in range(1,h-1):
            
            if (0<=ang[i,j]<22.5) or (157.5<=ang[i,j]<=180):
                q = img[i,j+1]
                r = img[i,j-1]
            elif 22.5<=ang[i,j]<67.5:
                q = img[i+1,j-1]
                r = img[i-1,j+1]
            elif 67.5<=ang[i,j]<112.5:
                q = img[i+1,j]
                r = img[i-1,j]
            elif 112.5<=ang[i,j]<157.5:
                q = img[i-1,j-1]
                r = img[i+1,j+1]
                
            if (img[i,j]>=q) and (img[i,j]>=r):
                blank_img[i,j] = img[i,j]
            else:
                blank_img[i,j] = 0
        
    return blank_img

def hysteresis_threshold(img,min_threshold_ratio,max_threshold_ratio,strong,weak):
    
    w,h = img.shape
    result_img = np.zeros((w,h),dtype=np.uint8)
    
    max_threshold = img.max()*max_threshold_ratio
    min_threshold = max_threshold*min_threshold_ratio
    
    strong_pixels_i,strong_pixels_j = np.where(img>=max_threshold)
    weak_pixels_i,weak_pixels_j = np.where((img>=min_threshold) & (img<=max_threshold))
    
    result_img[strong_pixels_i,strong_pixels_j] = strong
    result_img[weak_pixels_i,weak_pixels_j] = weak
    
    return result_img

def edge_tracking(img_1,strong,weak):
    
    img = img_1.copy()
    w,h = img.shape
    for i in range(1,w-1):
        for j in range(1,h-1):
            if img[i,j]==weak:
                
                if ((img[i,j-1]==strong or img[i-1,j-1]==strong) or (img[i-1,j]==strong) or (img[i-1,j]==strong) or 
                    (img[i-1,j+1]==strong) or (img[i,j+1]==strong) or (img[i+1,j+1]==strong) or (img[i+1,j]==strong)):
                    
                    img[i,j] = strong
                else:
                    img[i,j] = 0
    return img    
    

gauss_kernel = gaussian_kernel(3,3)
smooth_img = cv2.filter2D(img_gray, -1, gauss_kernel)
img_derivative,angle= sobel_filter(smooth_img)
supressed_img = non_max_supression(img_derivative,angle)
strong_pixels = np.int32(255)
weak_pixels = np.int32(25)
hysteresis_threshold_img = hysteresis_threshold(supressed_img,0.05,0.1,strong_pixels,weak_pixels)
edge_tracking_img = edge_tracking(hysteresis_threshold_img,strong_pixels,weak_pixels)

plt.subplot(231)
plt.imshow(img_gray,cmap = 'gray');
plt.axis("off")
plt.title('Original image');

plt.subplot(232)
plt.imshow(smooth_img,cmap = 'gray');
plt.axis("off")
plt.title('Smoothened image');

plt.subplot(233)
plt.imshow(img_derivative,cmap = 'gray');
plt.axis("off")
plt.title('Sobel filter');

plt.subplot(234)
plt.imshow(supressed_img,cmap = 'gray');
plt.axis("off")
plt.title('Non max\nsupression');

plt.subplot(235)
plt.imshow(hysteresis_threshold_img,cmap = 'gray');
plt.axis("off")
plt.title('Hysteresis\nthresholding');

plt.subplot(236)
plt.imshow(edge_tracking_img,cmap = 'gray');
plt.axis("off")
plt.title('Edge tracking');
plt.tight_layout(pad=1.0)

#Save image (Uncomment the line to save)
#plt.savefig("./Results/Result5.png",dpi=300,bbox_inches='tight')
```
